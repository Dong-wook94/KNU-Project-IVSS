```python
videos = []
download_dir = 'F:\project'
```

```python
youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey = DEVELOPER_KEY)
```

```python
next_token = ''
for i in range(10):
    if i == 0:
        search_response = youtube.search().list(
        q = search_q, 
        type = 'video',
        part = 'snippet',
        maxResults = 50).execute()
    else:
        search_response = youtube.search().list(
        q = search_q, 
        type = 'video',
        part = 'snippet',
        pageToken = next_token,
        maxResults = 50).execute()
        
    for search_result in search_response.get("items", []):
        videos.append(search_result['id']['videoId'])
    
    next_token = search_response.get('nextPageToken')
```

```python
videos[:300]
```

```python
url = "https://www.youtube.com/watch?v="
for v in videos[:200]:
    try:
        yt = YouTube(url + v)
        yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first().download(download_dir)
        print(url + v)
    except:
        print("XXX")
```
